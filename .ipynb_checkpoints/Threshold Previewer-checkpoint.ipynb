```python
def displayImages(img, title):
    fig, (ax1) = plt.subplots(1)
    fig.set_size_inches(10, 7)
    ax1.imshow(img)
    ax1.set_title(title)
```

```python
def adaptive_threshold_stack(input_image, block_size = 100, constant = -30):
    if block_size % 2 != 1:
        block_size += 1

    thresholded = []
    for i in range(input_image.shape[0]):
        thresholded.append(cv2.adaptiveThreshold(input_image[i], 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                                  cv2.THRESH_BINARY, block_size, constant)) # ADAPTIVE_THRESH_MEAN_C
    return np.array(thresholded)
```

```python
def updateAdapt(val):
    global z
    global block_size
    global constant
    global stack
    global ax2
    global ax
    global centers
    global gradient
    global gradient2
    global crop_s_XY
    global crop_e_XY
    thresholded = adaptive_threshold_stack(stack, int(block_size.val), int(constant.val))
    #thresholded = binarizeStack(stack)
    stackRGB = np.stack((stack, stack, stack), axis=-1)
    thresholdedRGB = np.stack((thresholded, np.zeros_like(thresholded),np.zeros_like(thresholded)), axis=-1)
    finalStack = stackRGB *0.6 + thresholdedRGB *0.4
    #ax2.imshow(stack[int(z.val)])
    # ax2.imshow(finalStack[int(z.val),crop_s_XY[0]:crop_e_XY[0], crop_s_XY[1]:crop_e_XY[1]]/255)
    ax2.imshow(finalStack[int(z.val),:,:]/255)
    
def updateHyst(val):
    global zHyst
    global low
    global high
    global stackHyst
    global ax2Hyst
    global axHyst
    global centersHyst
    global gradientHyst
    global gradient2Hyst
    global startIndex
    global movingAverage
    thresholded = hysteresisThresholdingStack(stackHyst, low.val, high.val)
    #thresholded = binarizeStack(stack)
    stackRGB = np.stack((stackHyst, stackHyst, stackHyst), axis=-1)
    thresholdedRGB = np.stack((thresholded, np.zeros_like(thresholded),np.zeros_like(thresholded)), axis=-1)
    finalStack = stackRGB *0.5 + thresholdedRGB *0.5
    #ax2.imshow(stack[int(z.val)])
    ax2Hyst.imshow(finalStack[int(zHyst.val)])

    axHyst.cla()
    axHyst.plot(centersHyst[startIndex:], movingAverage[0][startIndex:], color='black')
    axHyst.axvline(low.val, 0, 1, label='Low')
    axHyst.axvline(high.val, 0, 1, label='High')
```

```python
def chooseAdaptiveThresholding(img_stack, default_block_size=150, default_constant=-10):
    global z
    global block_size
    global constant
    global stack
    global ax2
    global ax
    global centers
    global gradient
    global gradient2
    global crop_s_XY
    global crop_e_XY
    ax = None
    ax2 = None
    # if np.max(img_stack) > 1:
    #     img_stack = img_stack / 255
    crop_s_XY = [400, 400]
    crop_e_XY = [1300, 1100]

    stack = img_stack

    BlockSizeInit = default_block_size
    ConstantInit = default_constant

    fig = plt.figure()
    fig.set_size_inches(10, 10)
    fig.tight_layout()
    # ax = plt.subplot(121)
    # plt.subplots_adjust(left=0.15, bottom=0.4)    

    # ax.grid(False)
    # ax.title('Choose hysteresis thresholds')
    # ax.xlabel('Intensity')
    # ax.ylabel('concentration')

    axcolor = 'lightgoldenrodyellow'
    axBlockSize = plt.axes([0.15, 0.3, 0.65, 0.03], facecolor=axcolor)
    axConstant = plt.axes([0.15, 0.35, 0.65, 0.03], facecolor=axcolor)
    axZ = plt.axes([0.15, 0.40, 0.65, 0.03], facecolor=axcolor)

    block_size = pylab.Slider(axBlockSize, 'BlockSize', 0, 250, valinit=BlockSizeInit, valstep = 1)
    constant = pylab.Slider(axConstant, 'Constant', -50, 50, valinit=ConstantInit, valstep = 1)
    z = pylab.Slider(axZ, 'z', 0, stack.shape[0] - 1, valinit=int(stack.shape[0]/2), valstep = 1)

    ax2 = plt.subplot(211)
    plt.subplots_adjust(hspace = .001)
    thresholded = adaptive_threshold_stack(stack, int(block_size.val), int(constant.val))
    #thresholded = binarizeStack(stack)
    stackRGB = np.stack((stack, stack, stack), axis=-1)
    thresholdedRGB = np.stack((thresholded, np.zeros_like(thresholded),np.zeros_like(thresholded)), axis=-1)
    finalStack = stackRGB *0.6 + thresholdedRGB *0.4
    #ax2.imshow(stack[int(z.val)])
    # ax2.imshow(finalStack[int(z.val),crop_s_XY[0]:crop_e_XY[0], crop_s_XY[1]:crop_e_XY[1]]/255)
    ax2.imshow(finalStack[int(z.val),:,:]/255)

    block_size.on_changed(updateAdapt)
    constant.on_changed(updateAdapt)
    z.on_changed(updateAdapt)

    plt.show()
```

```python
def hysteresisThresholdingStack(stackHyst, low=0.25, high=0.7):
    return apply_hysteresis_threshold(stackHyst, low, high) > 0

def determineHysteresisThresholds(img, outputPath=None, bins=256, movingAverageFrame=20, cutOffSlope=4, highVal=0.95):
    counts, centersHyst = histogram(img, nbins=bins)
    #remove 'black'
    counts = counts[1:]
    centersHyst = centersHyst[1:]

    # normalise
    # counts = counts/np.sum(counts)

    df = pd.DataFrame(counts)
    movingAverage = df.rolling(movingAverageFrame, center=True).mean()


    # first derivitive
    # gradient = []
    # gradient2 = []
    # for i in range(0,len(counts)-1):
    #     gradient.append(counts[i+1] - counts[i])
    
    # gradient.append(0)
    # df = pd.DataFrame(gradient)
    # movingAverage = df.rolling(movingAverageFrame, center=True).mean()


    #print(movingAverage)

    # calculate second derivitive
    # for i in range(0,len(movingAverage[0])-1):
    #     gradient2.append(movingAverage[0][i+1] - movingAverage[0][i])
    
    # gradient2.append(0)
    # plt.plot(centers[startIndex:], gradient[startIndex:], color='black')
    # plt.show()

    # df = pd.DataFrame(gradient2)
    # movingAverage = df.rolling(movingAverageFrame, center=True).mean()
    startIntensity = 10
    useIntensityLow = startIntensity
    useIntensityHigh = 0
    #for i in range(len(movingAverage[0])*3//4,startIntensity, -1):
    for i in range(len(movingAverage[0])*3//4,startIntensity, -1):
        # print(i,": ", movingAverage[0][i-10]/movingAverage[0][i+10] )
        #if movingAverage[0][i] <= mEdges:
        if movingAverage[0][i-10]/movingAverage[0][i+10] >= cutOffSlope:
              useIntensityLow = i
              print("Low intensity to be used: ", useIntensityLow/bins)
              print("High intensity to be used: ", (1.0-(1.0-useIntensityLow/bins)/2))
              #print("High intensity to be used: ", highVal)
              break  

    print(outputPath)
    if outputPath != None:
        plt.figure(figsize=(6, 4))
        plt.plot(centersHyst, counts, color='black') #movingAverage[0]
        plt.axvline(useIntensityLow/bins, 0, 1, label='Low', color="red")
        plt.axvline((1.0-(1.0-useIntensityLow/bins)/2), 0, 1, label='High', color="blue")
        plt.xlabel("Intensity")
        plt.ylabel("Count")
        plt.title("The automatically calculated hysteresis thresholding values")
        plt.tight_layout()
        #plt.axvline(highVal, 0, 1, label='High')
        plt.savefig(outputPath)
        print("Saved histogram")

    return (useIntensityLow/bins, (1.0-(1.0-useIntensityLow/bins)/2))
    #return (useIntensityLow/bins, highVal)

def chooseHysteresisParams(img, bins=256, remove_zero=False, remove_max=False):
    global zHyst
    global low
    global high
    global stackHyst
    global ax2Hyst
    global axHyst
    global centersHyst
    global gradientHyst
    global gradient2Hyst
    global startIndex
    global movingAverage
    axHyst = None
    ax2Hyst = None
    stackHyst = img
    startIndex = 0
    if remove_zero:
        startIndex = 1
    
    counts, centersHyst = histogram(img, nbins=bins)
    #remove 'black'
    counts = counts[1:]
    centersHyst = centersHyst[1:]

    df = pd.DataFrame(counts[startIndex:])
    movingAverage = df.rolling(20, center=True).mean()

    (lowInit, highInit) = determineHysteresisThresholds(img)

    axHyst = plt.subplot(121)
    plt.subplots_adjust(left=0.15, bottom=0.4)    

    # ax.grid(False)
    # ax.title('Choose hysteresis thresholds')
    # ax.xlabel('Intensity')
    # ax.ylabel('concentration')

    axcolor = 'lightgoldenrodyellow'
    axLow = plt.axes([0.15, 0.15, 0.65, 0.03], facecolor=axcolor)
    axHigh = plt.axes([0.15, 0.2, 0.65, 0.03], facecolor=axcolor)
    axZ = plt.axes([0.15, 0.25, 0.65, 0.03], facecolor=axcolor)

    low = pylab.Slider(axLow, 'Low', 0.0, 1.0, valinit=lowInit)
    high = pylab.Slider(axHigh, 'High', 0.0, 1.0, valinit=highInit)
    zHyst = pylab.Slider(axZ, 'z', 0, img.shape[0] - 1, valinit=2, valstep = 1)

    axHyst.plot(centersHyst[startIndex:], movingAverage[0][startIndex:], color='black')
    axHyst.axvline(low.val, 0, 1, label='Low')
    axHyst.axvline(high.val, 0, 1, label='High')

    ax2Hyst = plt.subplot(122)
    thresholded = hysteresisThresholdingStack(stackHyst, low.val*255, high.val*255)
    #thresholded = binarizeStack(stack)
    stackRGB = np.stack((stackHyst, stackHyst, stackHyst), axis=-1)
    thresholdedRGB = np.stack((thresholded, np.zeros_like(thresholded),np.zeros_like(thresholded)), axis=-1)
    finalStack = stackRGB *0.6 + thresholdedRGB *0.4
    #ax2.imshow(stack[int(z.val)])
    ax2Hyst.imshow(finalStack[int(zHyst.val)])

    low.on_changed(updateHyst)
    high.on_changed(updateHyst)
    zHyst.on_changed(updateHyst)

    plt.show()
```
